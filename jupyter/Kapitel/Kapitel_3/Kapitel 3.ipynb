## Die Entscheidung

```python
def entscheidung(summe):
    """Berechnung der Entscheidung zum Wert summe
    Input: summe
    Output: 1, falls summe >= 1
            0 sonst
    """
    if summe >= 1:
        return 1
    else:
        return 0
    
#-------------------------------------------------
#Berechnung der Entscheidung

ergebnis = entscheidung(1)

#Ausgabe in Zelle
print(ergebnis)
```

## Stufenfunktion

```python
# Import der Module
import matplotlib.pyplot as plt
# Ganz wichtig, sonst wird der Plot nicht angezeigt
%matplotlib inline

# x/y des Graphen
x = [-1,0,0.99,1,2]
# y Werte mit der Funktion entscheidung berechen
y = [entscheidung(i) for i in x]
# Erzeugen des Graphen, mit einer orangenen Stufe und Bezeichnung step
plt.step(x, y, color="orange", label='step')

# Die Achse setzen
plt.grid(True)
# Die horizontale und vertikale 0-Achse etwas dicker in Schwarz zeichen
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

# Achsenbeschriftung und -titel
plt.xlabel('Summe')
plt.ylabel('Ergebnis')
plt.title('Stufenfunktion')

# Legenden Platzierung festlegen
plt.legend(loc='lower right')

# Den Graphen anzeigen
plt.show()
```

## Vectoren

```python
# Import der Module
import matplotlib.pyplot as plt
# Ganz wichtig, sonst wird der Plot nicht angezeigt
%matplotlib inline

import numpy as np

# Die Funktion array wandelt eine Python-Liste in ein NumPy-Array um
# Die Koordinaten x1 = x
x1 = np.array([0,0,1,1])
# Die Koordinaten x2 = y
x2 = np.array([0,1,0,1])
# Die Farben für die Punkte
color = np.array(['black', 'black', 'red', 'black'])
# Die Punktgröße für jeden Punkt
size = np.array([100,100,500,100])

# Den Vektor x1 mit allen x1-Koordinaten ausgeben 
print('Der Vektor x1: ', x1)

# Die Achsen setzen
plt.grid(True)

# Den plot zeichnen für die x1- und x2-Koordinaten, Farbe und Größe
plt.scatter(x1,x2,c=color,s=size)

# Die Achsen setzen
plt.xlabel('x')
plt.ylabel('y')

# Die Achseneinteilung setzen
plt.xticks([0.0, 1.0])
plt.yticks([0.0, 1.0])

#Endlich den Plot ausgeben
plt.show()
```

## Iris-Datensatz Übungsaufgabe

```python
# Import der Module
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline

x1 = []
x2 = []
colors = []

dirname = os.path.dirname(os.path.abspath("__file__"))
filename = os.path.join(dirname, 'jupyter','Kapitel', 'Kapitel_3', 'iris.csv')


with open(filename, "r") as file:
    for line in file:
        values = line.split(',')
        
        if(values[0] == '\n'):
            continue
        
        x1.append(float(values[0]))
        x2.append(float(values[1]))

        if values[4] == 'Iris-setosa\n':
            colors.append('red')
        if values[4] == 'Iris-versicolor\n':
            colors.append('green')
        if values[4] == 'Iris-virginica\n':
            colors.append('blue')
            
# Die Achsen setzen
plt.grid(True)

# Den plot zeichnen für die x1- und x2-Koordinaten, Farbe und Größe
plt.scatter(np.array(x1),np.array(x2),c=np.array(colors))

# Die Achsen setzen
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('scatter-Plot')

#Endlich den Plot ausgeben
plt.show()
```

## Skalarprodukt mit NumPy

```python
# Die Multiplikation von Vektoren mit NumPy
import numpy as np

# NumPy-Array erzeugen
x = np.array([0,1])
# NumPy-Array erzeugen
w = np.array([0.5,0.7])

# Vektor x ausgeben
print("x = ", x)
# Vektor w ausgeben
print("w = ", w)

# NumPy-Arrays sind keine Matrizen, und
# die Operatoren *,+,-,/ funktionieren Element für Element
print("w*x =", w*x)

# Inneres Produkt. Man muss den Vektor nicht transponieren,
# das erledigt NumPy für uns
print("np.dot(w,x) =", np.dot(w,x))

# Alternative Syntax zum Dot-Product
print("w.dot(x) =", w.dot(x))
```
